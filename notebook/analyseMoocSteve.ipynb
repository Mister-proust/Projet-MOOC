```python
from pymongo import MongoClient

# Connexion à MongoDB
client = MongoClient("mongodb://localhost:27017/")

# Sélection de la base et de la collection
db = client["MOOC"]
collection = db["forum"]

# Petit test pour voir une donnée au hasard
doc = collection.find_one()
print(doc)
```

```python
# Nombre total de documents
nb_documents = collection.count_documents({})
print("Nombre total de documents :", nb_documents)
```

```python
# Est-ce qu’il manque des champs ?
# Vérifier si body (le texte principal) est bien là partout :
nb_missing_body = collection.count_documents({"content.body": {"$exists": False}})
print("Documents sans 'content.body' :", nb_missing_body)
# Vérifier la présence de course_id (le cours associé) :
nb_missing_course = collection.count_documents({"content.course_id": {"$exists": False}})
print("Documents sans 'content.course_id' :", nb_missing_course)
```

```python
# Quelles sont les valeurs distinctes ?
# Liste des cours (course_id) présents :
course_ids = collection.distinct("content.course_id")
print("Cours distincts :", course_ids)
print("Nombre de cours différents :", len(course_ids))
# Liste des courseware_title (nom des modules) :
courseware_titles = collection.distinct("content.courseware_title")
print("Modules distincts :", courseware_titles)
print("Nombre de modules différents :", len(courseware_titles))
```

```python
# Stat sur les commentaires :
# On récupère tous les threads (content.type == "thread")
threads = collection.find({"content.type": "thread"})

# Pour chaque thread, on prend la longueur de content.children (liste de commentaires)
total_comments = 0
total_threads = 0

for thread in threads:
    children = thread.get("content", {}).get("children", [])
    total_comments += len(children)
    total_threads += 1

# Calcul de la moyenne (attention à ne pas diviser par 0)
if total_threads > 0:
    average_comments = total_comments / total_threads
else:
    average_comments = 0

print(f"Nombre moyen de commentaires par fil : {average_comments:.2f}")
```
